## **Image Dataset Pre-Processing**
**Dataset from BlackBoard:** Facial Emotion Recognition dataset

**File:** facial-emotion-recognition.zip (432.768 MB)

- Reorganize image files into new subfolders sorted on category (8 facial expression)
- Image pre-processing (blob, hog, lbp)?

```python
import os
import shutil

# Define the root directory where the 19 folders are located
root_dir = "/Users/thomas/Desktop/IT3212/assignment_4/thomas/facial-emotion-recognition/images"

# Define the names of the target folders
target_folders = ['anger', 'contempt', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprised']

# Create target folders if they do not exist
for folder in target_folders:
    target_folder_path = os.path.join(root_dir, folder)
    if not os.path.exists(target_folder_path):
        os.makedirs(target_folder_path)

# Iterate through each of the 19 folders in the root directory
for folder_name in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder_name)

    # Skip if it's not a directory
    if not os.path.isdir(folder_path):
        continue

    # Process files in the current folder
    for file_name in os.listdir(folder_path):
        if file_name in ['Anger.jpg', 'Contempt.jpg', 'Disgust.jpg', 'Fear.jpg', 'Happy.jpg', 'Neutral.jpg', 'Sad.jpg', 'Surprised.jpg']:
            # Determine the target folder based on the file name
            if file_name == 'Anger.jpg':
                target_folder = 'anger'
            elif file_name == 'Contempt.jpg':
                target_folder = 'contempt'
            elif file_name == 'Disgust.jpg':
                target_folder = 'disgust'
            elif file_name == 'Fear.jpg':
                target_folder = 'fear'
            elif file_name == 'Happy.jpg':
                target_folder = 'happy'
            elif file_name == 'Neutral.jpg':
                target_folder = 'neutral'
            elif file_name == 'Sad.jpg':
                target_folder = 'sad'
            else:
                target_folder = 'surprised'
            
            # Construct the new file name with the suffix from the parent folder name
            new_file_name = f"{os.path.splitext(file_name)[0]}_{folder_name}{os.path.splitext(file_name)[1]}"
            
            # Define the source and destination paths
            source_file_path = os.path.join(folder_path, file_name)
            destination_file_path = os.path.join(root_dir, target_folder, new_file_name)
            
            # Move and rename the file
            shutil.move(source_file_path, destination_file_path)
            print(f"Moved: {file_name} -> {new_file_name} to {target_folder}")

# After moving all the files, remove the folders named from '0' to '18'
for i in range(19):
    folder_to_delete = os.path.join(root_dir, str(i))
    
    # Check if the folder exists and is empty
    if os.path.isdir(folder_to_delete):
        try:
            os.rmdir(folder_to_delete)  # Only removes empty directories
            print(f"Deleted empty folder: {folder_to_delete}")
        except OSError:
            print(f"Folder {folder_to_delete} is not empty or couldn't be deleted.")
```
