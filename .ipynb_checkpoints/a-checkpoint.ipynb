```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.neighbors import LocalOutlierFactor
```

```python
pd.set_option('display.float_format', '{:.3f}'.format)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False) 
```

```python
col = 'AxleWeight1'

df = pd.read_csv(
    './data/20221014-20 Kistler_R3_ostg(in).csv', 
    sep=';', 
    skiprows=6, 
    on_bad_lines='skip', 
    decimal='.',
    low_memory=False
).dropna(subset=[col])
```

```python
plt.hist(df[col], bins=100)
plt.title("Weight on the first axle, 1 week of measurements")
plt.xlabel("Weight, KG")
plt.ylabel("Registrations")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Assuming df is your DataFrame and col is the column you're clustering
dbscan = DBSCAN(eps=20)
outliers = dbscan.fit_predict(df[[col]])

# Append outlier labels to the DataFrame
df['outlier'] = outliers

# Create a figure with two subplots in one row
fig, ax = plt.subplots(1, 2, figsize=(12, 6))  # Adjust figsize as needed

# Scatter plot for DBSCAN results
scatter = ax[0].scatter(df.index, df[col], c=df['outlier'], cmap='viridis', marker='o', alpha=0.5)
# Highlight outliers
ax[0].scatter(df[df['outlier'] == -1].index, df[df['outlier'] == -1][col], color='red', label='Outliers')
ax[0].set_title("DBSCAN Clustering with Outliers Highlighted")
ax[0].set_xlabel("Index")
ax[0].set_ylabel("Weight, KG")
ax[0].legend()

# Optionally, keep only the non-outliers (outlier label != -1)
df_cleaned = df[df['outlier'] != -1]

# Histogram for cleaned data
ax[1].hist(df_cleaned[col], bins=100, color='blue', alpha=0.7)  # Adjust range as needed
ax[1].set_title("Weight on the first axle after oulier removal, 1 week of measurements")
ax[1].set_xlabel("Weight, KG")
ax[1].set_ylabel("Registrations")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
